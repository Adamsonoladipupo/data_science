```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/adamson/Downloads/master.csv")
df.head()
```

No. 1 

```python
population = df["population"]
plt.title("Population Histogram")
plt.hist(population)
plt.show()
```

2. Relationship between Population and Suicide Number 

```python
population = df["population"]
suicide = df["suicides_no"]
plt.scatter(population, suicide, color = "purple")
plt.show()
```

3. Bar chart of sex and population with color set to green

```python
sex = df["sex"]
population = df["population"]
plt.bar(sex, population)
plt.show()
```

Detailed number 3. labels, title and legends and use of grid

```python
sex = df["sex"]
plt.xlabel("Sex")
population = df["population"]
plt.ylabel("Population")
plt.title("Sex & Population Bar Chart")
plt.bar(sex, population)
plt.legend(title="Bar-Chart")
plt.grid()
plt.show()
```
